# Exploratory Data Analysis of Heart Disease Dataset

This notebook analyzes a publicly available heart disease dataset using
Python's Pandas and NumPy libraries. The analysis includes data exploration,
summary statistics, and correlation analysis.


```python
import pandas as pd
import numpy as np
```

```python
# Load the dataset
data = pd.read_csv("Heart_Disease_Prediction.csv")

# Display first few rows
data.head()
```

## Dataset Information

This section shows the structure of the dataset, including column names,
data types, and missing values.


```python
data.info()

```

## Summary Statistics

This section displays summary statistics such as mean, median, standard
deviation, minimum, and maximum values for numerical features.


```python
data.describe()
```

## Basic Statistical Measures

This section calculates mean, median, mode, and standard deviation
for each numerical feature in the dataset.


```python
data.mean(numeric_only=True)
```

```python
data.median(numeric_only=True)
```

```python
data.mode().iloc[0]
```

```python
data.std(numeric_only=True)
```

## Correlation Analysis

This section computes the correlation between numerical features
to identify relationships within the dataset.


```python
data.corr(numeric_only=True)
```

## Conclusion

In this analysis, we explored the heart disease dataset by examining its structure,
summary statistics, and basic statistical measures. Correlation analysis provided
insights into relationships between numerical features, which can be useful for
predictive modeling and further machine learning tasks.
