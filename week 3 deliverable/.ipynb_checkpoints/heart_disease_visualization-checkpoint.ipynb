# Data Visualization with Matplotlib and Seaborn

This notebook demonstrates basic data visualization techniques using
Matplotlib and Seaborn, including line plots, scatter plots, histograms,
and box plots.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv("Heart_Disease_Prediction.csv")

# Preview dataset
data.head()
```

## Line Plot using Matplotlib

A line plot is used to visualize trends in numerical data.


```python
plt.figure()
plt.plot(data.index, data["Age"])
plt.title("Line Plot of Age Distribution")
plt.xlabel("Index")
plt.ylabel("Age")
plt.grid(True)
plt.show()
```

## Scatter Plot using Matplotlib

A scatter plot is used to visualize the relationship between two variables.


```python
plt.figure()
plt.scatter(data["Age"], data["Cholesterol"])
plt.title("Scatter Plot of Age vs Cholesterol")
plt.xlabel("Age")
plt.ylabel("Cholesterol")
plt.grid(True)
plt.show()
```

## Histogram using Seaborn

Histograms show the distribution of a single numerical variable.


```python
plt.figure()
sns.histplot(data["Age"], bins=10, kde=True)
plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

## Box Plot using Seaborn

Box plots help identify data spread, median, and outliers.


```python
plt.figure()
sns.boxplot(y=data["Cholesterol"])
plt.title("Box Plot of Cholesterol Levels")
plt.ylabel("Cholesterol")
plt.show()
```

## Conclusion

In this notebook, Matplotlib was used to create line and scatter plots,
while Seaborn was used to visualize distributions through histograms
and box plots. These visualizations help in understanding trends,
relationships, and variability within the dataset.
